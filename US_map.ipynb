```python
import plotly.graph_objs as go 
from plotly.offline import init_notebook_mode,iplot,plot
init_notebook_mode(connected=True)
```

```python
df_confirmed_US = pd.read_csv('time_series_covid_19_confirmed_US.csv')
df_deaths_US = pd.read_csv('time_series_covid_19_deaths_US.csv')
```

```python
df_deaths_US.head()
```

```python
df_confirmed_US_1 = df_confirmed_US[df_confirmed_US['iso2']=='US']
df_confirmed_US_1 = df_confirmed_US_1.groupby('Province_State').sum().reset_index()

df_confirmed_US_1 = df_confirmed_US_1.drop(['UID','code3','FIPS','Lat','Long_'],axis=1)

df_confirmed_US_1 = df_confirmed_US_1[df_confirmed_US_1['Province_State']!= 'Diamond Princess']
df_confirmed_US_1 = df_confirmed_US_1[df_confirmed_US_1['Province_State']!='Grand Princess']
```

```python
state_codes = {
    'District of Columbia' : 'dc','Mississippi': 'MS', 'Oklahoma': 'OK', 
    'Delaware': 'DE', 'Minnesota': 'MN', 'Illinois': 'IL', 'Arkansas': 'AR', 
    'New Mexico': 'NM', 'Indiana': 'IN', 'Maryland': 'MD', 'Louisiana': 'LA', 
    'Idaho': 'ID', 'Wyoming': 'WY', 'Tennessee': 'TN', 'Arizona': 'AZ', 
    'Iowa': 'IA', 'Michigan': 'MI', 'Kansas': 'KS', 'Utah': 'UT', 
    'Virginia': 'VA', 'Oregon': 'OR', 'Connecticut': 'CT', 'Montana': 'MT', 
    'California': 'CA', 'Massachusetts': 'MA', 'West Virginia': 'WV', 
    'South Carolina': 'SC', 'New Hampshire': 'NH', 'Wisconsin': 'WI',
    'Vermont': 'VT', 'Georgia': 'GA', 'North Dakota': 'ND', 
    'Pennsylvania': 'PA', 'Florida': 'FL', 'Alaska': 'AK', 'Kentucky': 'KY', 
    'Hawaii': 'HI', 'Nebraska': 'NE', 'Missouri': 'MO', 'Ohio': 'OH', 
    'Alabama': 'AL', 'Rhode Island': 'RI', 'South Dakota': 'SD', 
    'Colorado': 'CO', 'New Jersey': 'NJ', 'Washington': 'WA', 
    'North Carolina': 'NC', 'New York': 'NY', 'Texas': 'TX', 
    'Nevada': 'NV', 'Maine': 'ME'}

df_confirmed_US_1['state_code'] = df_confirmed_US_1['Province_State'].apply(lambda x : state_codes[x])
```

```python
data = dict(type='choropleth',
            locations = df_confirmed_US_1['state_code'],
            locationmode = 'USA-states',
            colorscale = ["yellow","blue","green","violet","orange","red"],
            #text = df_confirmed_US_1['Province_State'],
            z = df_confirmed_US_1['4/2/20'],
            colorbar = {'title':'Confirmed'}
            )

layout = dict(title = 'Confirmed Cases in US',
              geo = dict(scope='usa')
             )

choromap = go.Figure(data = [data],layout = layout)
iplot(choromap)
```

```python
df_deaths_US_1 = df_deaths_US[df_confirmed_US['iso2']=='US']

df_deaths_US_1 = df_deaths_US_1.groupby('Province_State').sum().reset_index()

df_deaths_US_1 = df_deaths_US_1.drop(['UID','code3','FIPS','Lat','Long_','Population'],axis=1)

df_deaths_US_1 = df_deaths_US_1[df_deaths_US_1['Province_State']!= 'Diamond Princess']
df_deaths_US_1 = df_deaths_US_1[df_deaths_US_1['Province_State']!='Grand Princess']

df_deaths_US_1['state_code'] = df_deaths_US_1['Province_State'].apply(lambda x : state_codes[x])
```

```python
data = dict(type='choropleth',
            locations = df_deaths_US_1['state_code'],
            locationmode = 'USA-states',
            colorscale = ["yellow","blue","green","violet","orange","red"],
            z = df_deaths_US_1['4/2/20'],
            colorbar = {'title':'Deaths'}
            )

layout = dict(title = 'Deaths from Coronavirus in US',
              geo = dict(scope='usa')
             )

choromap = go.Figure(data = [data],layout = layout)
iplot(choromap)
```

```python
df_confirmed_US_2 = df_confirmed_US[df_confirmed_US['iso2']=='US']

df_confirmed_US_2 = df_confirmed_US_2.drop(['UID','code3','FIPS','Country_Region','iso3'],axis=1)

df_confirmed_US_2 = df_confirmed_US_2[df_confirmed_US_2['4/2/20']!= 0]
```

```python
data = go.Scattergeo(
        lon = df_confirmed_US_2['Long_'],
        lat = df_confirmed_US_2['Lat'],
        text = df_confirmed_US_2['4/2/20'],
        mode = 'markers',
        marker = dict(symbol = 'circle',size=5,colorscale = ["yellow","blue","green","violet","orange","red"]
                      ),
        marker_color = df_confirmed_US_2['4/2/20']
        )
layout = dict(title = 'Confirmed Cases',
              geo_scope = 'usa'
             )
choromap = go.Figure(data = [data],layout = layout)
iplot(choromap)
```
